```python
import pandas as pd
```

```python
el_universal = pd.read_csv('web_scrapper_curso_data_eng/eluniversal_2019_11_14_articles.csv', encoding = 'UTF-8')
#encoding = 'unicode_escape'  el encoding del video
el_universal
```

# data wrangling

```python
# 1. añadir el newpaper_uid al dataframe
# se agrega otra columna con su contenido

el_universal['newpaper_uid'] = 'eluniversal'

el_universal
```

```python
# 2. obtener el host
# apply genera transformaciones/operaciones personalizadas
#lambda es una función inline  ej: b = lambda argumentos : resultados
#netloc permite obtener el host

from urllib.parse import urlparse

el_universal['host'] = el_universal['url'].apply(lambda url: urlparse(url).netloc)

el_universal
```

```python
#obtenemos cada uno de los host y la cantidad
el_universal['host'].value_counts()
```

# Missing data

```python
# 3. Rellenar datos faltantes
#objetivo: identificar donde falta el titulo y rellenarlo con
#el titulo que viene en la url

missing_titles_mask = el_universal['title'].isna()

#el_universal[missing_titles_mask]['url'] va a elegir url's donde isna == true
#nos interesa la última parte de la url, ya que es la que contiene el titulo
#https://www.elgrafico.mx/la-roja/nina-de-14-an...
#(r'([^/]+)$) se define un conjunto que recorre todo lo que empieza con / (^/)
# pueden haber 1 o más "/" (+) 
# y obtine el último / hasta el final del string ($)
#(?P<missing_titles>[^/]+) otorga una etiqueta llamada missing_titles
# 

# los () en python sirven para concatenar funciones y
# esto lo que hace posible es tener las funciones en dif lineas
#applymap permite hacer una transformación
missing_titles = (el_universal[missing_titles_mask]['url']
                .str.extract(r'(?P<missing_titles>[^/]+)$')
                .applymap(lambda title: title.replace('-',' '))
                 )
#otra al ternativa para separar las palabras por - y luego hacer join con el espacio ' 'es:
#.astype(str).applymap(lambda title: title.replace('-',' '))) inclusive sin el .astype(str) funciona

"""
#otra alternativa que funciona
missing_titles = (el_universal[missing_titles_mask]['url']
                .str.extract(r'(?P<missing_titles>[^/]+)$')
                .applymap(lambda title: title.split('-'))
                .applymap(lambda title_word_list: ' '.join(title_word_list))
                #la parte de title_word_list no la entendí, de donde sale, es una palabra reservada?
                )
"""

el_universal.loc[missing_titles_mask, 'title'] = missing_titles.loc[:, 'missing_titles']
missing_titles
```

# adittional cleanup

```python
#4. add uid to rows --> add ids

#for this we will support in hashlib, usually is used in criptografy
# but in this case is used to generete unique number for url maping
# hash is a algorithms that receive a input(text-passwords)and return a alphanumeric output
import hashlib

#hashlib.md5 return one code of 128 bits, which parameters are bytes, encode is utf
#axis=1 for rows, = 0 for col
uids = (el_universal
           .apply(lambda row: hashlib.md5(bytes(row['url'].encode())), axis=1)
            #get hexadecimal representation
            .apply(lambda hash_object: hash_object.hexdigest())
           )

#add to col
el_universal['uid'] = uids
#we want new col are index, inplace = true is for update directly in table
el_universal.set_index('uid', inplace=True)

el_universal
```

```python
#delete \r and \n 
stripped_body = (el_universal
                .apply(lambda row: row['body'], axis=1) #get body
                .apply(lambda body: list(body)) #convert body in list letter
                .apply(lambda letters: list(map(lambda letter: letter.replace('\r', ''), letters))) #map(function, lista/iterable)
                .apply(lambda letters2: list(map(lambda letter2: letter2.replace('\n', ''), letters2)))
                .apply(lambda letters2: ''.join(letters2))
                 )

#stripped_body
el_universal['body'] = stripped_body
el_universal
```

# Data enrichment

```python
#6. tokenizar el titulo y el body
import nltk
#stopwords son palabras bastante utilizadasen en el lenguaje
#sin embargo no aportan nada ej (el la, los) 
from nltk.corpus import stopwords

#nltk.download('punkt')
#nltk.download('stopwords')

stop_words =set(stopwords.words('spanish'))

def tokenize_column(df, column_name):
    return (df
           .dropna()
            .apply(lambda row: nltk.word_tokenize(row[column_name]),axis=1)
            .apply(lambda tokens: list(filter(lambda token: token.isalpha(),tokens)))
            .apply(lambda tokens: list(map(lambda token: token.lower(),tokens)))
            .apply(lambda word_list: list(filter(lambda word: word not in stop_words, word_list)))
            .apply(lambda valid_word_list: len(valid_word_list))
           )
el_universal['n_tokens_title'] = tokenize_column(el_universal, 'title')
el_universal['n_tokens_body'] = tokenize_column(el_universal, 'body')

el_universal
```

# Duplicate values

```python
#6. Eliminar duplicados

#verificamos que cada unos de los titulos no tengan duplicados
el_universal['title'].value_counts()
```

```python
# a drop_duplicates le pasamos la columna donde vamos a eliminar duplicados
# first es para quedarnos con el primer duplicado
# e inplace es para remplazar 
el_universal.drop_duplicates(subset = ['title'], keep="first", inplace=True)
```
